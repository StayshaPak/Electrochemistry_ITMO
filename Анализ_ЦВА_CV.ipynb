```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Для поиска максимумов/минимумов

# 1. Загрузка данных из CSV-файла (или другого формата)
# Убедитесь, что ваш файл содержит столбцы "Potential" и "Current"
try:
    data = pd.read_csv("/content/drive/MyDrive/Магистратура/Д з на ДПО2025/Тема 2/Задание 1/Задание 1 - Задание 1.csv.csv")
except FileNotFoundError:
    print("Ошибка: Файл не найден.")
    exit()

# Убедитесь, что столбцы называются правильно (чувствительны к регистру)
if "Potential" not in data.columns or "Current" not in data.columns:
    print("Ошибка: Файл CSV должен содержать столбцы 'Potential' и 'Current'.")
    exit()

# 2. Построение графика
plt.figure(figsize=(10, 6))  # Установите размер графика для лучшей видимости
plt.plot(data["Potential"], data["Current"], color="blue")
plt.xlabel("Potential (V)")
plt.ylabel("Current (A)")  # Или Current (mA)
plt.title("Scan")
plt.grid(True)  # Добавление сетки для удобства чтения
```

```python
# Расчет анодного потенциала и тока
anodic_peak_index = data["Current"].idxmax()
anodic_potential = data["Potential"][anodic_peak_index]
anodic_current = data["Current"][anodic_peak_index]
```

```python
# Расчет катодного потенциала и тока
cathodic_peak_index = data["Current"].idxmin()
cathodic_potential = data["Potential"][cathodic_peak_index]
cathodic_current = data["Current"][cathodic_peak_index]
```

```python
# Вывод значений
print(f"Anodic potential, Ea = {anodic_potential:.3f} V")
print(f"Cathodic potential, Ec = {cathodic_potential:.3f} V")

print(f"Anodic current, Ia = {anodic_current:.8f} A")
print(f"Cathodic current, Ic = {cathodic_current:.8f} A")
```

```python
# 4. Расчет и вывод значений E1/2 и dE
E_half = (anodic_potential + cathodic_potential) / 2
delta_E = abs(anodic_potential - cathodic_potential)

print(f"E1/2 = {E_half:.5f} V")
print(f"Delta E = {delta_E:.6f} V")
```
